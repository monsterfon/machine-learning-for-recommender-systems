```python

import pandas as pd
import plotly.express as px
from ipywidgets import interact, widgets
from IPython.display import clear_output
```

```python
# Read the data
data = pd.read_csv('D:/apps/podatki/FEATURES_5s/FEATURES_5s/S2_C.csv')

# Create a time column
data['Time'] = data.index

# List of signal columns (excluding non-signal columns)
non_signal_cols = ['Time', 'Target']
signal_cols = [col for col in data.columns if col not in non_signal_cols]

```

```python
data.dropna(inplace=True)
print(data.shape)
print(data.columns)
print(data.head())
```

```python
for col in ['Time'] + signal_cols:
    print(f"{col}: {data[col].notnull().sum()} non-null values")
```

```python
# Interactive plot function

def plot_signals(signals):
     
    if not signals:
        print("No signals selected.")
        return
    if not isinstance(signals, list):
        signals = [signals]
    

    # Copy the data to avoid modifying the original DataFrame
    plot_data = data.copy()
    # Normalize the 'Target' variable to match the scale of the signals
    plot_data['Normalized_Target'] = plot_data['Target'] / plot_data['Target'].max() * plot_data[signals].max().max()
    # Add 'Normalized_Target' to the list of signals
    signals.append('Normalized_Target')
    fig = px.line(plot_data, x='Time', y=signals, title='Interactive Time Series Plot')
    fig.show()



```

```python

# Interactive selector for plotting signals
interact(plot_signals, signals=signal_cols)
```
