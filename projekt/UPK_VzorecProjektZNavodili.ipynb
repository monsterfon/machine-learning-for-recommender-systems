# Povzetek

V povzetku na kratko povzamemo, o čem naš seminar govori. Pomagamo si s
povzetkom in zaključkom iz našega gradiva (članka).

Seminarska naloga obravnava standard MPEG-7, formalno poimenovan
“Multimedia Content Description Inter-face,” ki se uporablja za
opisovanje vsebin z meta podatki. To uporabniku omogoča učinkoviteje in
uspešneje iskanje ter pridobivanje takšnih vsebin, kot z današnjimi
tekstovnimi brskalniki. Naloga je smiselno razdeljena na tri sklope, od
katerih se prvi del ukvarja z uvodom v standard MPEG-7. Temeljni del
vsebuje pregled in opis tehničnih značilnosti standarda. Ta del se
zaključi s pregledom združljivosti standarda in prikazom primerov
različnih aplikaciji pri uporabi standarda MPEG-7. Zaključni del naloge
je kratek povzetek kaj nam standard prinaša in kje ga bomo v prihodnje
zasledili.

Stil pisanja: ne pišemo v 1. osebi in dobesedno, kot da smo mi avtorji
članka. Pišemo spis o članku (oziroma temi), npr: »Avtorji so v članku z

Ne prevajamo članka dobesedno, ampak smiselno povzamemo s svojimi

**Metode, algoritmi, delovanje :** ne spuščamo se v podrobnosti

**Evaluacija sistema** : tukaj opišemo načine testiranja uspešnosti

ta sistem deluje v primerjavi z drugimii na istem področju.

**Primeri uporabe, zaključek.**

The Moving Picture Coding Experts Group (MPEG) je skupina iz Ženeve
ISO/IEC (International Standards Organization/International
Electro-technical Committee), ki skrbi za razvoj internacionalnih
standardov kompresije, dekompresije videa in avdia ter kombinacijo le
teh. Na kratko bom predstavil, za kaj se uporabljajo posamezni standardi
in v čem se standard MPEG-7 razlikuje od drugih. Ta organizacija je
razvila oziroma še razvija naslednje standarde:

► MPEG-1: Namenjen kodiranju avdio in video skupaj ali posamič.

► MPEG-2: Se uporablja za digitalno televizijo, namenjen satelitskemu
oddajanju in televizijski industriji pri prehodu iz analognega v
digitalne formate.

► MPEG-4: Napreden format za kompresijo, posebej prilagojen za učinkovit
prenos videa preko Interneta.

► **<u>MPEG-7:</u>** Je razvijajoč se standard, namenjen opisovanju
multimedijskih objektov. Z razliko od ostalih ni namenjen kodiranju in
kompresiji videa.

► MPEG-21: Je standard, na katerem organizacija še dela. Namenjen bo
dodeljevanju pravic in omejitev digitalnih vsebin od avtorja do
uporabnika.

## Predstavitev problema

Z multimedijskimi vsebinami se srečujemo vsaki dan. Dan za dnem je na
spletu na razpolago vse več in več avdiovizualnih informacij iz
različnih virov po celem svetu, katere se uporablja za različne namene.
Vzemimo kot primer iskanja sliko motorja, ki ga je vozil Arnold
Schwarzenegger v filmu Terminator 2, da bi poiskali podobne motorje na
spletu. Tu pa se pojavi vprašanje kako, naj sistem ve o čem govori film.
Takoj lahko ugotovimo, da je največji problem pri multimedijskih
vsebinah razkorak med digitalnim zapisom slike in zvoka in med pomenskim
zapisom, ki pove kaj predstavljajo podatki. Ta problem lahko odpravimo
tako, da uporabimo metapodatke in z njimi opišemo vsebino dokumenta v
digitalni obliki, tako da jo odjemalec na podlagi opisa najde. Dandanes
so metapodatki o vsebinah potrebni zaradi kompleksnosti dela z
digitalnimi vsebinami. Obstajajo različne profesionalne organizacije, ki
se ukvarjajo z standardi metapodatkov. Sam se bom skozi nalogo
osredotočil in predstavil standard MPEG-7. Nekaj najpogosteje
uporabljenih standardov:

-   **<u>MPEG-7</u>**

-   SMPTE Metadata Dictionary;

-   Dublin Core;

-   EBU P/Meta;

-   TV Anytime;

-   Podcast

# Vsebina : Standard MPEG-7

## Uvod v standard MPEG-7

**MPEG-7** je razvijajoč se standard, namenjen opisovanje multimedijskih
vsebin z meta podatki, ki uporabniku omogoča učinkoviteje in uspešneje
iskanje in pridobivanje takšnih vsebin, kot z današnjimi tekstovnimi
brskalniki. Ni namenjen kodiranju videa, temveč je le način kako opisati
elemente v multimedijskem toku, tako da do njih dostopamo preko baze. Ko
govorimo o opisovanju se ne nanašamo na kodiranje samega »filma«, ampak
mislimo o elementih, ki v njem nastopajo. Podobno, kot pri tekstovnem
besedilu, kjer lahko poizvedujemo po posameznih besedah, lahko pri tem
poizvedujemo po objektih, ki nastopajo v njem. Standard prinaša
najbogatejši nabor avdio vizualnih opisov s pomočjo meta podatkov. Opis
zajema tri dele avdio vizualnih vsebin:

► katalog ( naslov, avtorja, pravice),

► semantiko ( kdo, kaj, kdaj, kje, informacije o predmetih in dogodkih)
in

► strukturne značilnosti ( barvni histogram slike, posneti značilen zvok
določenega instrumenta)

MPEG-7
ne bo standardiziral izvlečke oziroma povzetke Avdio Vizualnih opisov,
niti ne bo določil iskalnike ali programe, ki bi lahko uporabil takšne
opise. Prepustil bo kreativnost družbam iskalnikov, da bodo tovrstne
informacije njihovi brskalniki znali iskati ter pridobivati.

Ko govorimo o standardu MPEG-7 lahko izpostavim še to, da opis standarda
ni nujno prisoten v istem računalniku ali na isti mreži skupaj z
originalnimi »stream podatki«, ampak se lahko nahaja nekje drugje. V tem
primeru pa so potrebni mehanizmi za sinhronizacijo med opisom in
originalnim medijem.

## Tehnične značilnosti standarda MPEG-7

Med najbolj pomembno lastnost standarda MPEG-7 štejemo to, da je
združljiv z različnimi programi in okolji. To so zagotovili tako, da so
uporabili fleksibilno ogrodje za opisovanje avdio vizualnih podatkov.
Elemente standarda je potrebno stabilizirat tako, da opis dokumenta
ustreza XML-u, hkrati pa vsebuje lastnosti standarda in upošteva opise
strukture. Standard za svoje opise uporablja XML jezik, ki je bil izbran
zato, ker je jezik popularen, uporaben, enostaven, v svojem vzponu in
predvsem zato ker je XML združljiv z veliko drugimi standardi kot so
SMPTE Metadata Dictionary, Dublin Core, EBU P/Meta,... To so zagotovili
tako, da so definirali multimedijske knjižnice metod in orodij. Temelje
standarda predstavljajo naslednji elementi:

**► opisnik (D)** - definira sintakso in semantiko od vsake značilnosti
(meta podatkov)

**► opisna shema (DS)** - definira strukturo in semantiko odnosa med
opisniki in opisnimi shemami

**► opisna definicija jezika (DDL)** - je eden pomembnejših delov
jezika, ki nam dovoljuje ustvarjanje opisnih shem in opisnikov, hkrati
pa omogoča upravljanje in spreminjanje že obstoječih opisnih shem

**► sistemska orodja** - nam omogočajo razmnoževanje opisov,
usklajevanje opisov z vsebino, nudi nam mehanizme in kodne predstavitve
v binarnem in tekstovnem zapisu, za uspešno hrambo, prenos, upravljanje
ter zaščito lastnine.

# Zaključek

Zaključek iz članka, kaj so ugotovitve iz članka glede uporabnosti
sistemov, ter tudi naše ugotovitve po raziskavi tega problema.

MPEG-7 je standard za opisovanje multimedijskih vsebin s pomočjo meta
podatkov, vsebuje bogato opisno orodje, s katerim si pomagamo pri
opisovanju multimedijskih vsebin. Najbolj pomembno pri MPEG-7 standardu
je to, da je splošen. To pomeni, da je združljiv z zelo veliko programi
in okolji. Je fleksibilno ogrodje za opisovane AV vsebin. Postal je
standard leta 2001 in je še zmeraj v razvijanju, tako da lahko
pričakujemo v bližnji prihodnosti iskanje AV vsebin tako hitro, kot
danes iščemo po internetu.
