## 1. Uvod <a class="anchor" id="uvod"></a>

Podatki poskusa WESAD vključujejo več vrst fizioloških signalov (npr. EDA, EKG, dihanje, temperatura, gibanje), zbranih v različnih afektivnih stanjih (npr. stres, meditacija). Te meritve odražajo trenutne fiziološke odzive posameznika in omogočajo vpogled v dogajanje v telesu ob različnih dražljajih. Pri analizi signala so ključne ustrezne značilke, kot sta na primer povprečje ali standardni odklon, saj pomagajo pri razumevanju, kako se posameznikovo fiziološko stanje spreminja skozi čas. Z dodatno vizualizacijo pridobimo boljši vpogled v vzorce signalov, kar je posebej uporabno pri opazovanju sprememb ob različnih fazah poskusa (npr. sproščena faza, obremenitev), kjer lahko jasno prikažemo prehode med stanji in ocenimo njihovo intenziteto ali trajanje.

### 1.1. Opredelitev področja <a class="anchor" id="opredelitev-podrocja"></a>



WESAD je odprtokodna podatkovna množica, ki združuje fiziološke in gibalne podatke (npr. BVP, ECG, EDA, EMG, dihanje, temperatura in pospešek) za prepoznavanje stresa in afektivnih stanj. Meritve so bile opravljene pri 15 udeležencih s senzorji, nameščenimi na prsni in zapestni del telesa, v treh različnih stanjih: nevtralno, stresno in stanje zabave. Avtorji so po eksperimentu izvedli tudi samoevalvacijo (ankete o stresu, čustvih itn.) in pripravili osnovni nabor značilk, ki ga lahko uporabniki te množice razširijo ali prilagodijo. Prve ocene učinkovitosti napovedi stanja s standardnimi metodami strojnega učenja so dosegle do 80-odstotno natančnost (pri razločevanju treh stanj) in do 93-odstotno natančnost (pri razločevanju stresa od ostalih stanj). WESAD služi kot izhodišče za raziskave na področju analize stresa in napovedovanja afektivnih stanj z nosljivimi napravami.

### 1.2. Predstavitev problema <a class="anchor" id="predstavitev-problema"></a>

Aplikativna postavitev problema na področju projekta WESAD se osredotoča na razvoj vizualizatorja za analizo in interpretacijo množice različnih fizioloških signalov (EDA, temperatura, dihanje, gibanje, EKG itd.). Cilj je omogočiti uporabnikom intuitiven vpogled v posamezne signale ter njihove izračunane značilke. Vizualizator je ključnega pomena za primerjavo značilk med različnimi osebami in afektivnimi stanji (stres, meditacija itd.), saj omogoča identifikacijo vzorcev in trendov, ki so bistveni za razumevanje dinamike fizioloških odzivov. S pomočjo interaktivnih grafov in vizualnih prikazov lahko raziskovalci lažje prepoznajo pomembne korelacije in odstopanja, kar pripomore k boljšemu oblikovanju napovednih modelov ter izboljšanju uporabniške izkušnje.

## 2. Teoretično ozadje in trenutno stanje <a class="anchor" id="teoreticno-ozadje-in-trenutno-stanje"></a>

"Affective computing" se ukvarja s prepoznavanjem in interpretacijo čustvenih stanj posameznika na podlagi različnih podatkovnih virov. Ključni cilj je zaznavanje stresa, ki ima pomembne posledice za dobrobit posameznika in medsebojno interakcijo človek-računalnik. Za natančno identifikacijo čustvenih stanj se uporabljajo različni fiziološki signali, kot so elektrodermalna aktivnost (EDA), srčni utrip (HR), elektrokardiogram (EKG), elektromiogram (EMG) ter drugi. Analiza teh signalov in njihovih značilnosti, kot so povprečje in standardna deviacija, omogoča razvoj modelov za napovedovanje afektivnih stanj, kar prispeva k boljši prilagojeni komunikaciji in podpori uporabnikom v realnem času.

V zadnjem času je na voljo več podatkovnih množic za raziskovanje zaznavanja stresa in afektivnih stanj, med katerimi je ena izmed najpomembnejših WESAD (Wearable Stress and Affect Detection) dataset. Ta množica podatkov vključuje multimodalne in visokokakovostne podatke iz različnih senzorjev, nameščenih na zapestju in prsih, zbranih od 15 subjektov med laboratorijskimi poskusi.  Z uporabo teh podatkov raziskovalci razvijajo in primerjajo modele za obdelavo signalov, ekstrakcijo značilnosti ter napovedovanje stresa in drugih afektivnih stanj, kar prispeva k napredku na področju nosljivih tehnologij in interakcije človek-računalnik.

### 2.2. Algoritmi <a class="anchor" id="algoritmi"></a>


### Kako deluje `interact`

Funkcija `interact` iz knjižnice **ipywidgets** samodejno generira uporabniške vmesnike (kot so drsniki, spustni seznami itd.) za vsak argument v funkciji, ki jo ovija. Potrebno je določiti:

1. Funkcijo, ki jo želite poklicati (npr. `plot_boxplots`).
2. Parametre (argumente) te funkcije skupaj z veljavnimi vrednostnimi razponi ali seznami.
3. Ko premaknete drsnik ali izberete drugo vrednost iz spustnega seznama, `interact` ponovno pokliče funkcijo z novimi parametri.

Na primer:




<p style="margin-bottom:2cm;"></p>
<img src="https://raw.githubusercontent.com/andrejkk/ORvTK_SlidesImgs/master/footer_full.jpg">

```python

from ipywidgets import interact

def example_function(x, text):
    print(f"x: {x}, text: {text}")

interact(
    example_function,
    x=(0, 10),          # creates a slider from 0 to 10
    text=["option1","option2"]  # creates a dropdown menu
)
```

### 2.3. Merjenje uspešnosti rešitve <a class="anchor" id="merjenje-uspesnosti-resitve"></a>

### Merjenje uspešnosti rešitve – Prikaz rezulatov na grafih

Pri merjenju uspešnosti rešitve je ključno, da pridobljene vrednosti in metrike  ustrezno vizualiziramo. S prikazovanjem grafov lahko hitro opazimo morebitna odstopanja ali vzorce, ki niso takoj razvidni iz surovih številskih podatkov.


- **Histogram**: Prikazuje porazdelitev podatkov in je uporaben za analizo frekvence vrednosti.
- **Boxplot**: Prikazuje porazdelitev podatkov in morebitne odstopajoče vrednosti, kar omogoča primerjavo med različnimi skupinami.
- **Linechart**: Prikazuje podatke v obliki črte, kar omogoča spremljanje trendov skozi čas.

Z ustrezno izbiro tipa grafa uporabniku ponudimo takojšen vpogled v delovanje rešitve in njeno učinkovitost.
